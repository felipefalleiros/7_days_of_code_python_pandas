### Criar uma nova coluna com os valores da localização, para refletir a respectiva classe geral na CDU

```python
def classificar_localizacao(x):
    if 0 <= x <= 99:
        return 'Generalidades. Ciência e conhecimento'
    elif 100 <= x <= 199:
        return 'Filosofia e psicologia'
    elif 200 <= x <= 299:
        return 'Religião'
    elif 300 <= x <= 399:
        return 'Ciências sociais'
    elif 400 <= x <= 499:
        return 'Classe vaga. Provisoriamente não ocupada.'
    elif 500 <= x <= 599:
        return 'Matemática e ciências naturais.'
    elif 600 <= x <= 699:
        return 'Ciências aplicadas.'
    elif 700 <= x <= 799:
        return 'Belas artes'
    elif 800 <= x <= 899:
        return 'Linguagem. Língua. Linguística.'
    else:
        return 'Geografia. Biografia. História.'

df_merged['classe_geral'] = df_merged['localizacao'].apply(classificar_localizacao)
```

### A diretoria também gostaria de gerenciar melhor os recursos humanos da biblioteca de acordo com a demanda de trabalho existente. Por exemplo:


- gerenciar a programação de férias dos colaboradores de acordo com os meses de menor demanda

- programar atividades que não sejam de atendimento ao usuário para períodos específicos de menor demanda.


 ### Gere uma tabela com a quantidade total de exemplares emprestados por mês e descubra quais meses são os que possuem a maior quantidade de empréstimos realizados. Plote um gráfico de linhas.

```python
df_merged['mes'] = df_merged['data_emprestimo'].dt.month
df_merged.head()
```

```python
df_emprestimos_por_mes = df_merged.groupby('mes')['id_emprestimo'].count().to_frame()
df_emprestimos_por_mes
```

```python
meses = {
    1: 'Janeiro', 2: 'Fevereiro', 3: 'Março', 4: 'Abril',
    5: 'Maio', 6: 'Junho', 7: 'Julho', 8: 'Agosto',
    9: 'Setembro', 10: 'Outubro', 11: 'Novembro', 12: 'Dezembro'
}

df_emprestimos_por_mes.index = df_emprestimos_por_mes.index.map(meses)
df_emprestimos_por_mes.rename(columns={'id_emprestimo':'total'}, inplace=True)
df_emprestimos_por_mes
```

```python
meses = df_emprestimos_por_mes.index

fig, ax = plt.subplots(figsize=(14,8))
ax.plot(meses, df_emprestimos_por_mes['total'])
ax.set_title('Quantidade de exemplares emprestados por mês', fontsize=20)
ax.set_xlabel('Meses', fontsize=16)
ax.set_ylabel('Empréstimos', fontsize=16)

plt.grid(True)
```

Visto que a maior parte do público das bibliotecas são estudantes principalmente de graduação e pós-graduação, afim de um melhor gerenciamento do RH da biblioteca o atendimento ao público tem uma menor demanda nos meses de férias como janeiro, julho e dezembro, como podemos observar no gráfico acima. Logo esses meses são os ideais para programar as férias de seus colaboradores que participam ativamente de atividades de atendimento ao público.

### Verificar quais foram os horários com maior quantidade de empréstimos ao longo de um dia inteiro.

```python
df_merged['hora_emprestimo'] = df_merged['data_emprestimo'].dt.hour
df_merged.head()
```

```python
df_emprestimos_por_hora = df_merged['hora_emprestimo'].value_counts().sort_index().to_frame()
df_emprestimos_por_hora = df_emprestimos_por_hora.reset_index()
df_emprestimos_por_hora
```

```python
import matplotlib.pyplot as plt
import numpy as np

horas = df_emprestimos_por_hora['hora_emprestimo']
total = df_emprestimos_por_hora['count']

# Normalizando (%) os valores de 'count' para definir as cores
norm = plt.Normalize(df_emprestimos_por_hora['count'].min(), df_emprestimos_por_hora['count'].max())
colors = plt.cm.Blues(norm(df_emprestimos_por_hora['count']))

# Definindo as posições das barras com espaçamento extra
x_positions = np.arange(len(horas)) * 1.2  # Multiplica por 1.2 para adicionar espaço entre as barras

# Criando o gráfico de barras com espaçamento
plt.bar(x_positions, total, color=colors, width=0.8)  # A largura pode ser ajustada conforme necessário

# Ajustando os xticks para corresponder às novas posições com os rótulos formatados
plt.xticks(ticks=x_positions, labels=[f'{h}' for h in horas])

# Adicionando título e rótulos aos eixos
plt.title('Total de Empréstimos por Hora')
plt.xlabel('Hora do Empréstimo')
plt.ylabel('Total de Empréstimos')

plt.tight_layout()  # Ajusta automaticamente o layout para evitar sobreposição
plt.show()
```

Através do gráfico podemos observar que os horários de maior pico são 10h e 16h. Nós horários de início da manhã as 7h e início da noite as 19h são os melhores horários para se fazer atividades além do atendimento ao público por serem os horários com menor movimento.

```python
def quantidade_percentual(variavel):
    df = pd.DataFrame(df_merged[variavel].value_counts())
    df.columns = ['quantidade']
    df['percentual'] = round((df['quantidade'] / df['quantidade'].sum()) *100, 1)

    return df
```
