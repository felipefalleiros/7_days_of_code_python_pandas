### **A quantidade de empréstimos está aumentando ou diminuindo ao decorrer dos últimos anos?**

```python
# Cria uma nova coluna chamada ano extraindo apenas o ano da data de emprestimo
df_merged['ano'] = df_merged['data_emprestimo'].dt.year
```

```python
# Conta a quantidade de vezes que cada valor de ano aparece no df e exibindo na ordem cornológica através do sort.index()
df_merged['ano'].value_counts().sort_index()
```

```python
import matplotlib.pyplot as plt
import numpy as np
anos = df_merged['ano'].unique()
qtd_emprestimos_por_ano = df_merged['ano'].value_counts().sort_index()

fig, ax = plt.subplots(figsize=(10,8))
ax.plot(anos, qtd_emprestimos_por_ano)
ax.set_title('Quantidade de empréstimos de livros de 2010 a 2020', fontsize=20)
ax.set_xlabel('anos', fontsize=16)
ax.set_ylabel('Empréstimos', fontsize=16)

plt.yticks(np.arange(20000, 240000, 20000))
plt.grid(True)
```

Através dos dados acima podemos perceber que houve um aumento na quantidade de empréstimos de 2010 a 2013, depois teve uma oscilação entre 2014 e 2017 e por fim uma queda entre 2018 e 2020, tendo tem 2020 uma queda brusca na quantidaade de empréstimos de livros.
Essa queda brusca pode ser motivada pela maior procura de conteúdos online.

### Criar uma nova coluna com os valores da localização, para refletir a respectiva classe geral na CDU

```python
def classificar_localizacao(x):
    if 0 <= x <= 99:
        return 'Generalidades. Ciência e conhecimento'
    elif 100 <= x <= 199:
        return 'Filosofia e psicologia'
    elif 200 <= x <= 299:
        return 'Religião'
    elif 300 <= x <= 399:
        return 'Ciências sociais'
    elif 400 <= x <= 499:
        return 'Classe vaga. Provisoriamente não ocupada.'
    elif 500 <= x <= 599:
        return 'Matemática e ciências naturais.'
    elif 600 <= x <= 699:
        return 'Ciências aplicadas.'
    elif 700 <= x <= 799:
        return 'Belas artes'
    elif 800 <= x <= 899:
        return 'Linguagem. Língua. Linguística.'
    else:
        return 'Geografia. Biografia. História.'

df_merged['classe_geral'] = df_merged['localizacao'].apply(classificar_localizacao)
```

### Alterar o formato da coluna 'matricula_ou_siape' para string para melhorar a legibilidade

```python
# realiza o cast de tipo dos dados em que o valor da coluna matricula_ou_siape não são nulos
# Faz primeiro a conversão para inteiro, eliminando as casas decimais e depois para str
df_merged['matricula_ou_siape'] = df_merged['matricula_ou_siape'].apply(lambda x: str(int(x)) if pd.notnull(x) else x)
```
